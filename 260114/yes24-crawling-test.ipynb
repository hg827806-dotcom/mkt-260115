- 200 : 클라이언트 요청에 서버가 문제없이 통신했다
- 300 : 클라이언트의 요청에 서버가 통신을 했으나 다른 곳으로 우회시켰다 (응답은 했으나 다른 곳으로 보냄)
- 400 : 클라이언트 요청이 잘못된 경우 -> 서버에서 정상적으로 응답할 수 없음 -> ex) 404. 사이트의 주소를 확인할 수 없습니다
- 500 : 클라이언트의 요청에는 문제 없으나, 서버에서 운행을 정상적으로 할 수 없을 경우 -> ex) 수강신청 시 서버 폭발할때
