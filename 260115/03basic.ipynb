# 데이터 자료구조 1 - 리스트 (list)
 - 문자열, 정수, 실수, 논리형 -> (변수 : 값) : 변수 당 값 1나 -> 그럼 300개의 값이 필요하면 300개의 변수가 필요.
 - 변수 : 여러가지의 값을 그룸화
   - 공통 분모를 가지고 있는 값들의 모임 -> 특성을 분류할 수 있다.
   - 상대적으로 컴퓨터 메모리 공간 효율적 사용 가능
   - 대부분의 크롤링한 결과값 -> 리스트 형태의 자료 구조 값이 결정됨
 - 변수명 = [값, 값, 값, 값] : 리스트의 형식
 - 리스트 & 문자열 : 유사성을 띄고 있다. -> 비슷한 문법 구조를 가지고 있다. (같지는 x)
 - 리스트, 문자열 모두 공통 속성을 가지고 있기 때문에 위와 같은 말이 이야기가 나오는 것이다. : '__iter__' 를 둘 다 가짐
 - '__iter__' -> iterable : 반복할 수 있는 / index 가지고 있을 수 밖에 없다.
 - 리스트 값을 추가하는 방법
   - append() : 리스크에 값을 추가할 때 사용 (뒤에서 부터 값을 추가)
   - 리스트 + 리스트 연산 : 하나의 통으로 묶어버림
   - insert (추가하고자하는 인덱스, 넣고자 하는 값)
   - extend (여러가의 값을 한 번에 넣고자 할 때)
 - 값을 제거하는 방법
   - pop() : 리스트에서 값을 제거 (뒤에서) /-> 인자값으로 인덱스 값 부여 가능 -> 해당 인덱스의 값이 제거
   - remove() : 리스트에서 특정 값을 직접 선택해서 제거 가능
   - del 리스트명 [인덱스] : 해당 인덱스 값을 제거
 - 슬라이싱 기능 = 문자열과 동일 작동
 - 리스트 자료 구조 안에 입력될 값의 타입이 상이해도 가능!

```python
location = ["서울시", "경기도", "인천시"]

print(location)
print(location[0]) # 인덱스를 가지고 있다.
print(location[1])
print(location[2])
print(location[-1])

# print(type(location)) -> 결과값이 <class 'list'>라고 알려줌 / 여기서 type은 이 함수가 어떤 타입의 함수인지 알려줌
```
