# 기초문법
- 출력 : print() -> 내장함수 = built in function / 내장 : 애초에 파이썬 문법 탄생될 때부터 존재 -> 내장된 프로그램 / 문자만 가능한 것이 아님, 정수나, 사칙연산도 가능
  
- 입력 : input("당신의 이름은?") -> 사용자로부터 어떤 값을 입력받을 수 있음 / placeholder = 입력해야할 값의 정보
  - 해당함수를 통해서 수집한 값은 무조건 문자열이 된다는 사실 기억
    
- 연산 :
  - 사칙연산 + @ -> + - * / %(나머지) **(제곱근) -> '+' : 문자간의 연결 연산자로서도 사용가능하다!
  - 비교연산 : <(작다), >(크다), <=(작거나 크다), >=(크거나 작다),  ==(왼쪽과 오른쪽이 같다),  !=(왼쪽과 오른쪽이 다르다 /  (좌항 = 우항, 우항의 값을 좌항으로 삽입, 적용 / 같다가 아님)
    
- 변수 :
  - variable -> 변할 수 있는 값을 넣는 공간 = 그릇
  - 변수명 작명 -> 직관, 명시 -> 누가봐도 한눈에 알아볼 수 있는 변수명 / 숫자가 먼저올 수 없음 (중간에 오는건 가능) ex) num01=0 / 01num=x
  - 팁 : 스네이크 표기법(뱀) -> case_study / 카멜 표기법(낙타) -> caseStudy
    
- 자료형 : 변수안에 담을 수 있는 값의 형태가 무엇이냐? -> 문자열(String) "" VS 문자 (Character) '' (파이썬에선 문자열이든 문자든 "", '' 상관없다) / 정수(Int) / 실수(float) / 논리형(Boolean) = 자료들
  - 형변환 -> 자동형변환 : input() (자동적으로 변환) / 명시적형변환 : (수동적으로 형태를 변환)
  - 자료용 확인 : 
  - int() : 정수형 변환
  - float() : 실수형 변환
  - str() : 문자열 변환
  - bool() : 논리형 변환
  - list() : 리스트 자료구조 변환
  - dict() : 딕셔너리 자료구조 변환
  - tulie() : 튜플 자료구조 변환
  - set() : 세트(집합) 자료구조 변환

- 문자열 : 특별한 기능이 있음 -> 독립적인 함수 혹은 속성들이 많이 존재
  - count() : 특정 문자열에서 어떤 단어가 몇 번 사용되었는지를 집계하고자 할 떄 (대, 소 문자 구분)
  - len() : length = 길이
  - find() : 해당 문자열에서 인자값으로 전달된 문자가 몊번때 인덱스에 위치하는지 찾고자 할 때
  - replace() : 특정 문자열을 대체하고자 할 때
  - index : 문자열 안에 각각의 문자는 고유한 숫자(메모리 공간의 주소)를 가지고 있다. -> 문자열의 왼쪽부터 숫자 시작 : 0부터 시작
  - slicing : 문자열 안에서 특정 구간을 잘라내서 가져오고 싶을 때 / 인덱스를 알아야한다.
  - strip() : 특정 문자열 앞, 뒤쪽에 여백 혹은 여백으로 인지할 수 있는 escape sequence (\n, \t) 문자열을 제거하고자 할 때
  - format() : 특정 문자열을 원하는 위치에 출력하고자 할 때, 사용할 수 있는 기능 (.메서드 함수)

- 형식 지정자
  - 특정 타입의 값을 출력하고자 할 때 어떤 형태로 값을 찾아와서 출력 결정
  - %s - string : 문자열
  - %c - character : 문자
  - %d - int : 정수형
  - %f - float : 실수형

```python
text = "banana"

# idx = text.find("a" , 3) # 3반째에 있는 a를 찾고싶어
index = text.find("a") # 1, 3, 5

next_index = text.find("a" , index +1) # 다음번의 a / 만일 5번때를 찾고 싶다면 3+1사용
print(index, next_index)

# print(dir(text)) # find 가 있어서 사용 가능
```

```python
# 반복문 사용

text = "banana"
target = "a"


# idx = text.find("a") # 결과값이 -1이 나옴 : 존재하지 않은 문자이기 떄문임 
# 예외 조항 처리 : 어떤 명령문을 실행시키고자 할 떄, 만약 ~가 아니라면, 해라!!

idx = text.find(target)

while idx != -1 :
    print(idx)
    idx = text.find(target, idx +1)
idx


```

```python
text = "python"

print(text)
print(text[0]) # 해달 문자의 0번째 값에 있는 인덱스 찾을때
print(text[1]) # 해달 문자의 1번째 값에 있는 인덱스 찾을때

# 어떤 문자열이든, 마지막번떄 문자의 인덱스 넘버는 전체 문자열 개수 -1과 동일하다.

print(len(text)) # text의 총 문자 갯수가 멸개인가?
print(len(text) -1) # text의 마지막 문자열
print(text[len(text) -1]) # 마지막 문자열이 무엇인가? [] 사용

print(text[-1])
print(text[-2])

# 특정사이트 접속 > 리뷰 크롤링 > 댓글 :




```

```python
text = "python"

print(text)
print(text[2])
print(text[2:4]) # 찾아오고자하는 문자열[start : finish+1] -> 2번 3번의 문자열을 잧고 싶을 때.

print(text[2:6])
print(text[2:]) # 슬라이싱에서 마지막 번째 인덱스의 생략은 곧 끝까지를 의미한다. 
print(text[2:len(text)]) # 해당 구문을 많이 사용함. 
print(text[2:-1]) # 선생님!!



```

```python
name = "I have a {}, I have an {}"

print(name)
print(name.format("pen", "apple")) # 순서대로 들어감
print("I have a {}, I have an {}".format("pen", "apple"))
print("I have a {1}, I have an {0}".format("pen", "apple"))
# 인덱스랑 연결됨  / 메서드 함수(method)함수 : 앞에 객체라는 함수가 있어야 한다. (함수 앞에 온점이 있나 없나로도 구분 가능)
# 문자열 형식의 객체. format = 이 문법의 사용 -> 메서드 함수
# 만약 그냥 format을 사용한다면 그냥 독립적으로 움직이는 함수로 다른 결과값이 나온다. -> (문자o, 숫자 x) -> 소수점의 위치를 일치시키거나 할 때 사용.
```

```python
# 파이썬 내장함수 가운데, 소수점 자리를 올리거나, 내리는 독립 기능은 없다.
# 이런 경우에는 외부에서 라이브러리 모듈을 찾아와 사용해야함
# 수학객체 라이브러리 모듈

import math

interest = 0.891
result01 = math.floor(interest) # floor : 바다 -> 버리다
print(result01)

result02 = math.ceil(interest) # ceil : 천장 -> 올리다
print(result02)

```

```python
# print("I have a {}, I have an {}".format("pen", "apple")) # 팬과 애플은 문자열


print("I have a %s, I have an %s" % ("pen", "apple")) # 팬과 애플은 문자열
print("I have a %f, I have an %s" % (3.14, "apple")) # 3.14는 실수 애플은 문자열
print("I have a %f, I have an %d" % (3.14, 7)) # 3.14는 실수 7은 정수


print("I have a %f, I have an %c" % (3.14, 'A')) # 엄격한 프로그래밍 언어 -> "", ''
print("I have a %f, I have an %s" % (3.14, 'AB')) # 문자가 아닌 문자열이기 때문에 %C (X) / %S (0)


print("I have a %f, I have an %c" % (3.14, 65)) # ASCII Chart를 보면 알 수 있듯이 문자와 정수는 서로 치환이 된다. 여기서 컴퓨느텉 65를 문자 A로 인색했기 때문에 가능하다.




```
